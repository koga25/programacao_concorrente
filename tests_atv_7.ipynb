```python

import subprocess as subp
import matplotlib.pyplot as plt
import numpy as np
import json
import time
```

```python
file_names = ["test_json_0_PC.txt", "test_json_1_PC.txt", "test_json_5_PC.txt", 
              "test_json_10_PC.txt", "test_json_100_PC.txt", "test_json_1000_PC.txt", 
              "test_json_10000_PC.txt"]
durations = []
```

```python
for file_name in file_names:
    txt_file = open(f"./Atividade_7/{file_name}", 'r')
    time_list = txt_file.readlines()
    text_lines = len(time_list)
    time_list = list(map(int, time_list))
    durations.append(round(sum(time_list)/text_lines))
    
```

```python
durations
```

```python
x = np.arange(len([0, 1, 5, 10, 100, 1000, 10000]))  # the label locations
width = 0.2  # the width of the bars

durations_bar = plt.bar(x , durations, width, label='pub/sub')

plt.xticks(x, [0, 1, 5, 10, 100, 1000, 10000])
plt.xlabel("Prefetch count")
plt.ylabel("Média do tempo de publicação entre 10000 \nrequisições (ms)")
plt.title("Diferenças de performance entre tamanhos diferentes do prefetch count")
plt.legend()

plt.bar_label(durations_bar, padding=3)

plt.savefig("./atv_7_graphs.png")
plt.show()
```
